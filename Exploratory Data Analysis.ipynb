```python
corr=data.corr()
```

```python
print corr
```

```python
import seaborn as sns

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(5, 5))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.4,
            square=True, xticklabels=True, yticklabels=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
```

```python
data.boxplot(column=['srch_adults_cnt', 'srch_children_cnt','srch_rm_cnt', 'cnt'], by='is_booking')
plt.ylim((0,4))
```

```python
plt.hist(data['srch_rm_cnt'])
```

```python
plt.hist(data[data['is_booking']==0]['srch_rm_cnt'], color='b', label='Not Booked')
plt.hist(data[data['is_booking']==1]['srch_rm_cnt'], color='r', alpha=0.5, label='Booked')
plt.title('Room Count Histogram by booked vs. not booked')
plt.legend()
plt.show()
```
